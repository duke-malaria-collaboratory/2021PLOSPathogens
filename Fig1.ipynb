```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline

plt.figure(figsize=(20,30))
```

```python
df = pd.read_csv('Data/PfEMP1TimeCourse.csv')
```

```python
sns.set_context("paper", font_scale=1)
fig, axs = plt.subplots(2, 3,figsize=(8.5, 8.5))

#fig.suptitle('3D7 $var$ gene transcription', fontsize=20)
sns.lineplot(
    data=df[df['ABC']=='A'],
    x='HPI', 
    y='TPM', 
    hue= 'ORF',
    style = 'Genotype',
    palette = 'Set2',
    ax=axs[0,0]
    ) 
axs[0,0].title.set_text('Group A $var$ genes')
axs[0,0].legend(fontsize='5')

sns.lineplot(
    data=df[df['ABC']=='B'],
    x='HPI', 
    y='TPM', 
    hue= 'ORF',
    style = 'Genotype',
    palette = 'Set2',
    ax=axs[0,1]
    )
axs[0,1].title.set_text('Group B $var$ genes')
axs[0,1].legend(fontsize='4')

sns.lineplot(
    data=df[df['ABC']=='C'],
    x='HPI', 
    y='TPM', 
    hue= 'ORF',
    style = 'Genotype',
    palette = 'Set2',
    ax=axs[0,2]
    )
axs[0,2].title.set_text('Group C $var$ genes')
axs[0,2].legend(fontsize='5')

sns.lineplot(
    data=df[df['ABC']=='BA'],
    x='HPI', 
    y='TPM',
    hue= 'ORF',
    style = 'Genotype',
    palette = 'Set2',
    ax=axs[1,0]
    )
axs[1,0].title.set_text('Group BA $var$ genes')
axs[1,0].legend(fontsize='5')

sns.lineplot(
    data=df[df['ABC']=='BC'],
    x='HPI', 
    y='TPM',
    hue= 'ORF',
    style = 'Genotype',
    palette = 'Set2',
    ax=axs[1,1]
    )
axs[1,1].title.set_text('Group BC $var$ genes')
axs[1,1].legend(fontsize='5')

sns.lineplot(
    data=df[df['ABC']=='Var2'],
    x='HPI', 
    y='TPM',
    hue= 'ORF',
    style = 'Genotype',
    palette = 'Set2',
    ax=axs[1,2]
    )
axs[1,2].title.set_text('Var2CSA')
axs[1,2].legend(fontsize='5')
plt.tight_layout()
plt.savefig('Figs/Fig1.png')
plt.savefig('Figs/Fig1.pdf')
```
